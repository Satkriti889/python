```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
```

```python
df=pd.read_csv('ai_job_dataset.csv')

df.head()
```

```python
df["experience_level"].value_counts()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
graph=sns.countplot(x="experience_level" , data=df,hue="experience_level")
plt.title("Experience")
plt.xlabel("Experience")
plt.ylabel("Count")

for container in graph.containers:
    graph.bar_label(container)
plt.show()
```

```python
category_count=df["experience_level"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(category_count, labels=category_count.index, autopct='%.2f%%')
plt.title('Pie Chart of Job Experience')

plt.legend(labels=category_count.index, title="Experience", loc="upper right")
plt.show()

```

```python
category_count=df["salary_currency"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(category_count, labels=category_count.index, autopct='%.2f%%')
plt.title('Pie Chart of salary currency')

plt.legend(labels=category_count.index, title="Currency", loc="upper right")
plt.show()

```

```python
sns.scatterplot(data=df, x="experience_level", y="salary_usd", color="blue")
plt.title('Experience vs Salary')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(18,8))

sns.histplot(df['salary_usd'], bins=np.arange(df['salary_usd'].min(),df['salary_usd'].max(),5),color='steelblue',edgecolor='black',kde=True)
plt.title('Distribution of salary usd')
plt.xlabel('salary_usd')
plt.ylabel('Count')
plt.show()
```
