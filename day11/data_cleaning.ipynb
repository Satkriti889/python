#  Data Cleaning with Pandas

This notebook covers essential data cleaning tasks using the `pandas` library in Python. Tasks include:

- Handling null/missing values
- Removing duplicates
- Converting and correcting data types
- Basic consistency checks

```python

import pandas as pd

# Load Titanic dataset (use URL)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Display first few rows
df.head()
```

```python
filter_df= df[df['Name'].notnull()]
filter_df
```

```python
df.dropna(subset=['Name'],inplace=True)
df
```

```python
df.dropna(subset=['Embarked'],inplace=True)
df
```

```python
df.dropna(subset='Name',axis=1,inplace=True)
df
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df1=df['Age'].fillna(df['Age'].mean(), inplace=True)
print(df1)
```

```python
df.fillna({'Age':df['Age'].mean()},inplace=True)
```

Parameter	             Behavior
drop_duplicates()	     Removes duplicate rows
keep='first' (default)	 Keeps the first occurrence of duplicates
keep='last'	             Keeps the last occurrence
keep=False	             Drops all duplicates
subset=['col1','col2']	 Checks duplicates only in specified columns
